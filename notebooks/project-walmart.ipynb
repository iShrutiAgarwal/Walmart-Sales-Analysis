```python
# importing dependencies
import pandas as pd

#mysql toolkit
import pymysql #this will work as an adapter to connect to mysql database
from sqlalchemy import create_engine #this will help us to create a connection to the database
```

```python
print(pd.__version__)
```

```python
df = pd.read_csv("walmart.csv", encoding_errors="ignore")
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#all duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# dropping all rows with missing records
df.dropna(inplace=True)
df.isnull().sum()
df.shape
```

```python
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)
df.head()
```

```python
df.info()
```

```python
df['total_price'] = df['quantity'] * df['unit_price']
df.head()
```

```python
# mysql connection
engine_mysql = create_engine("mysql+pymysql://root:i_Shruti18@localhost:3306/walmart_db")

try:
    engine_mysql
    print("Connection to MySQL database successful!")
except:
    print("Connection to MySQL database failed.")
```

```python
df.to_csv("walmart_cleaned_data.csv", index=False)
```
